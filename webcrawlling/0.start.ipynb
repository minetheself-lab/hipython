html속에 있는 요소를 꺼내는 방법
1. 요소를 직접 지정 > find("요소명", id="id명")
2. css의 요소 검색 방법 > select("선택자")

선택자란, css에서 요소를 지정하는 방법이다.

css에서 #은 id, .은 클래스. 예를 들어 "#subject" ".contents"
자식은 >로 지정. "span>b"
여러개 요소 목록은 , 예를 들어 'div,hi"

- select()는 지정한 선택자의 요소를 모두 가져온다. [리스트로]
- select_one()은 지정한 선택자 요소중 첫번째 요소만 반환.

```python
b_tag=soup.select("div>b")
print(b_tag[0].text)
```

```python
h1 = soup.find("h1")
print(h1.get_text())
```

```python
subject = soup.find("div", id="subject")
print(subject.get_text())
```

```python
contents = soup.find("div", class_="contents")
print(contents.get_text())
```

```python
third_div = soup.find_all("div")[2]
b_inside = third_div.find("b")
print(b_inside.get_text())
```

```python
b_bold=soup.select('span>b')
print(b_bold[0].text)
```
