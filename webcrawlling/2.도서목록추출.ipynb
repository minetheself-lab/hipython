# 객체로 다루기

```python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    def __str__(self):
        return f"{self.title} / {self.author} / {self.price}"
```

```python
# 페이지 지정
base_url = 'https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageNumber='
books=[]
for page in range(1, 7):
    url = f'{base_url}{page}'
    bestseller = requests.get(url)
    soup = BeautifulSoup(bestseller.text, 'html.parser')
    items = soup.select(selector)
    
    for i, item in enumerate(items):
        title = item.select_one(title_selector).text.strip()
        author = item.select_one(author_selector).text.strip()
        price = item.select_one(price_selector).text.strip()
        book = Book(title, author, price)
        books.append(book)
        print(f"{(page-1)*24 + i+1}. {book}")
```

```python
import sqlite3
conn = sqlite3.connect('_books.db')
cursor = conn.cursor()

# 테이블 생성
cursor.execute('''
CREATE TABLE IF NOT EXISTS books (
    rank INTEGER PRIMARY KEY,
    title TEXT,
    author TEXT,
    price INTEGER
)
''')
```

```python
# 데이터 삽입
for book in books:
    cursor.execute('''
    INSERT INTO books (title, author, price) VALUES (?, ?, ?)
    ''', (book.title, book.author, book.price))

conn.commit()
cursor.close()
conn.close()
```
