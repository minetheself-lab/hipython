```python
user_input=input('정수의 목록을 공백으로 구분해서 입력하세요. 종료는 q')
str_int=user_input.split()
numbers=[]

for num in str_int:
    if num !='0':
        numbers.append(int(num))
print(numbers)  
    
```

```python
while True:
    user_input=input('정수의 목록을 공백으로 구분해서 입력하세요. 종료는 q')
    if user_input=="q":
        break
    str_int=user_input.split()
    numbers=[] #None은 초기값 안줄거야라는 뜻, []리스트 타입 예약

    for num in str_int:
        if num !='0':
            numbers.append(int(num))
    print(numbers)  
    if numbers==[]:
        print('입력한 숫자가 없습니다.')
    plus_numbers=[]
    minus_numbers=[]
    for num in numbers:
        if num>0:
            plus_numbers.append(num)
        else:
            minus_numbers.append(num)
    pls_cnt=len(plus_numbers)
    minus_cnt=len(minus_numbers)
    total=sum(numbers)
    print(f'양수:{pls_cnt}개 음수:{minus_cnt}개 총 합계:{total}')
    #print(plus_numbers, minus_numbers)
```
