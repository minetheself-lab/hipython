```python
sentence=input("문장을 입력하세요:")
words=["파이썬", "코딩", "데이터", "AI"]


found_word=None
for word in words:
    if word in sentence:
        found_word=word
        break

if found_word:
    print(f"입력한 문장에서 '{found_word}' 단어를 찾았습니다'")
else:
    print("입력한 문장에 해당 단어가 없습니다.")
```

```python
#가위바위보 게임
player='바위'
computer='바위'
result='승' if((player=='가위' and computer=='보') or\
             (player=='바위' and computer=='가위') or\
              (player=='보' and computer=='바위')) else \
        '무승부' if player==computer else\
        '패'
print (result)
```

```python
# 재고관리
inventory = [
    '노트북',
    '마우스',
    '키보드',
    '모니터'
]
# 0 품절, 10< 부족, <50 적정, >50 충분
for name, qty in inventory.items():
#qty = inventory['노트북']
    if qty >50 :
        status = '충분'
    elif qty >10 :
        status = '적정'
    elif qty > 0 :
        status = '부족'
    else :
        status = '품절'
    print(f'{name} : {qty} 로 {status} 상태입니다.')
```
