```python
#간식박스 생성함수
def make_snackbox(): #값을 받지 않고 함수안에서 정해진 값으로 하겠다.
    snackbox=['초코파이', '쿠키']
    return snackbox[0] 
aaa=make_snackbox()
print(aaa)
```

```python
#간식박스 추가함수
def add_snackbox(qqq): # 뭔지 모르지만 밖에서 입력받은 값으로 함수안에서 써먹겠다. 아무이름이나 넣어도 됨
    make_msg=input('먹고 싶은 간식을 추가하세요. 쉼표(,)로 연결하세요.')
    qqq.extend(make_msg)
    return qqq
add_snackbox(aaa)
print (aaa)
```
