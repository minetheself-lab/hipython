```python
#실습) 대한민국 남성의 평균 몸무게는 70kg 이다. 맞는지 틀리는지 가설 검정
```

```python
import numpy as np
from scipy import stats
```

```python
np.random.seed(123)
weight=np.random.uniform(40,100,100)
weight [:5]
```

```python
test_result, p_value=stats.ttest_1samp(weight, 70)
```

```python
if p_value < 0.05:
    print ('귀무가설 기각')
    print ('대한민국 남성의 평균 몸무게는 70kg이 아니다.')
else :
    print('귀무가설 채택')
    print('대한민국 남성의 평균 몸무게는 70kg이다.')
```
