```python
#My SQL 서버에 연결
con = pymysql.connect(
  host='localhost',
  user='root',
  password = '1234',
  db= 'wntrade',
  charset = 'utf8'
  )
#커서생성
cursor = con.cursor()
```

```python
#sql쿼리 실행
sql = """
select 도시, avg(마일리지) as 평균마일리지, (select avg(마일리지)
					from 고객
					) as 전체평균마일리지
from 고객 
group by 도시
having avg(마일리지) > (select avg(마일리지)
					from 고객
					);
"""
```

```python
cursor.execute(sql)
result=cursor.fetchall()
```

```python
result
```

```python
for row in result:
  print(row)
```

```python
cursor.close()
con.close()
```
