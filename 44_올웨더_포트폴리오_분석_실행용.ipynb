## STEP 1: ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°

````python
# í•„ìˆ˜ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜ (ì²˜ìŒ ì‹¤í–‰ ì‹œ)
# !pip install pykrx pandas plotly

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from pykrx import stock
from datetime import datetime

# ì¶œë ¥ ì˜µì…˜ ì„¤ì •
pd.options.display.float_format = '{:.2f}'.format

print("âœ… ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¡œë”© ì™„ë£Œ!")
````

## STEP 2: í¬íŠ¸í´ë¦¬ì˜¤ êµ¬ì„±í•˜ê¸°

**ìì‚° ë°°ë¶„:**
- ì£¼ì‹ 30% (ì„±ì¥)
- ì¥ê¸°ì±„ê¶Œ 40% (ì•ˆì •ì„±)
- ì¤‘ê¸°ì±„ê¶Œ 15%
- ì›ìì¬ 7.5%
- ê¸ˆ 7.5%

````python
portfolio_df = pd.DataFrame({
    'ìì‚°êµ°': ['ì£¼ì‹', 'ì¥ê¸°ì±„ê¶Œ', 'ì¤‘ê¸°ì±„ê¶Œ', 'ì›ìì¬', 'ê¸ˆ'],
    'ë¹„ì¤‘(%)': [30, 40, 15, 7.5, 7.5],
    'ETFëª…': [
        'KODEX 200',
        'ACE ë¯¸êµ­30ë…„êµ­ì±„ì•¡í‹°ë¸Œ', 
        'TIGER ë¯¸êµ­ì±„10ë…„ì„ ë¬¼',
        'KODEX ê³¨ë“œì„ ë¬¼(H)',
        'KODEX ê³¨ë“œì„ ë¬¼(H)'
    ],
    'í‹°ì»¤': ['133690', '411060', '356540', '130680', '261240']
})

print("ğŸ“‹ í¬íŠ¸í´ë¦¬ì˜¤ êµ¬ì„±")
print("=" * 70)
print(portfolio_df.to_string(index=False))
print(f"\nâœ… ì´ ë¹„ì¤‘: {portfolio_df['ë¹„ì¤‘(%)'].sum()}%")
````

## STEP 3: ìì‚° ë¹„ì¤‘ ì‹œê°í™”

````python
fig = px.pie(
    portfolio_df,
    values='ë¹„ì¤‘(%)',
    names='ìì‚°êµ°',
    title='ì˜¬ì›¨ë” í¬íŠ¸í´ë¦¬ì˜¤ ìì‚° ë°°ë¶„',
    hole=0.3,
    color_discrete_sequence=px.colors.qualitative.Set3
)

fig.update_traces(
    textposition='inside',
    textinfo='percent+label'
)

fig.show()
````

## STEP 4: ETF ë°ì´í„° ìˆ˜ì§‘

â° **ì£¼ì˜:** 5ê°œ ETF ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ë¯€ë¡œ 10~15ì´ˆ ì •ë„ ì†Œìš”ë©ë‹ˆë‹¤.

````python
# ë¶„ì„ ê¸°ê°„ ì„¤ì • (ì›í•˜ëŠ” ê¸°ê°„ìœ¼ë¡œ ë³€ê²½ ê°€ëŠ¥)
start_date = '20230101'
end_date = '20260131'

print(f"ğŸ“¥ {start_date} ~ {end_date} ê¸°ê°„ì˜ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤...\n")

price_data = {}

for idx, row in portfolio_df.iterrows():
    ticker = row['í‹°ì»¤']
    asset_name = row['ìì‚°êµ°']
    
    print(f"  â”” {asset_name} ({ticker}) ìˆ˜ì§‘ ì¤‘...", end=" ")
    
    try:
        df = stock.get_etf_ohlcv_by_date(start_date, end_date, ticker)
        price_data[asset_name] = df['ì¢…ê°€']
        print(f"âœ… ì™„ë£Œ ({len(df)}ê°œ)")
    except Exception as e:
        print(f"âŒ ì˜¤ë¥˜: {e}")

price_df = pd.DataFrame(price_data)

print(f"\nâœ… ì „ì²´ ë°ì´í„° ìˆ˜ì§‘ ì™„ë£Œ! (ì´ {len(price_df)}ì¼)")
print("\nğŸ“Š ìµœê·¼ 5ì¼ ì¢…ê°€:")
print(price_df.tail())
````

## STEP 5: ë°ì´í„° í’ˆì§ˆ í™•ì¸

````python
print("ğŸ” ë°ì´í„° í’ˆì§ˆ ì²´í¬")
print("=" * 70)
print(f"ê¸°ê°„: {price_df.index[0]} ~ {price_df.index[-1]}")
print(f"ê±°ë˜ì¼ìˆ˜: {len(price_df)}ì¼")
print(f"\nê²°ì¸¡ì¹˜ ê°œìˆ˜:")
print(price_df.isnull().sum())

if price_df.isnull().sum().sum() > 0:
    print("\nâš ï¸  ê²°ì¸¡ì¹˜ë¥¼ ì´ì „ ê°’ìœ¼ë¡œ ì±„ì›ë‹ˆë‹¤.")
    price_df = price_df.fillna(method='ffill')

print("\nâœ… ë°ì´í„° ì¤€ë¹„ ì™„ë£Œ!")
````

## STEP 6: ì›”ë³„ ìˆ˜ìµë¥  ê³„ì‚°

**ìˆ˜ìµë¥  ê³µì‹:**
```
ìˆ˜ìµë¥  = (ì´ë²ˆ ë‹¬ ê°€ê²© - ì§€ë‚œ ë‹¬ ê°€ê²©) / ì§€ë‚œ ë‹¬ ê°€ê²©
```

````python
# ì›”ë§ ê°€ê²© ì¶”ì¶œ
print("ğŸ“… ì›”ë§ ê°€ê²© ì¶”ì¶œ")
monthly_price = price_df.resample('M').last()
print(f"  ì¼ë³„ {len(price_df)}ê°œ â†’ ì›”ë³„ {len(monthly_price)}ê°œë¡œ ë³€í™˜\n")

# ì›”ë³„ ìˆ˜ìµë¥  ê³„ì‚°
monthly_return = monthly_price.pct_change().fillna(0)

print("ìµœê·¼ 3ê°œì›” ì›”ë§ ê°€ê²©:")
print(monthly_price.tail(3))

print("\nìµœê·¼ 3ê°œì›” ìˆ˜ìµë¥  (%):")
print((monthly_return.tail(3) * 100).round(2))
````

## STEP 7: í¬íŠ¸í´ë¦¬ì˜¤ ì „ì²´ ìˆ˜ìµë¥  ê³„ì‚°

**ê°€ì¤‘ í‰ê· :** ê° ìì‚° ìˆ˜ìµë¥  Ã— ë¹„ì¤‘ì„ ëª¨ë‘ ë”í•˜ê¸°

````python
# ë¹„ì¤‘ì„ ì†Œìˆ˜ë¡œ ë³€í™˜
weights = portfolio_df.set_index('ìì‚°êµ°')['ë¹„ì¤‘(%)'] / 100

print("âš–ï¸  ê° ìì‚°ì˜ ê°€ì¤‘ì¹˜:")
print(weights)

# í¬íŠ¸í´ë¦¬ì˜¤ ìˆ˜ìµë¥  ê³„ì‚°
portfolio_return = (monthly_return * weights).sum(axis=1)

print("\nğŸ“Š í¬íŠ¸í´ë¦¬ì˜¤ ì›”ë³„ ìˆ˜ìµë¥  (ìµœê·¼ 6ê°œì›”):")
recent_returns = portfolio_return.tail(6) * 100
for date, ret in recent_returns.items():
    emoji = "ğŸ“ˆ" if ret > 0 else "ğŸ“‰" if ret < 0 else "â¡ï¸"
    print(f"  {emoji} {date.strftime('%Y-%m')}: {ret:+.2f}%")
````

## STEP 8: ëˆ„ì  ìˆ˜ìµ ê³„ì‚° (ë³µë¦¬ íš¨ê³¼)

**ë³µë¦¬:** ë§¤ì›” ìˆ˜ìµì„ ê³„ì† ê³±í•˜ê¸°!

````python
# ëˆ„ì  ìˆ˜ìµ ê³„ì‚°
portfolio_value = (1 + portfolio_return).cumprod()

print("ğŸ’ í¬íŠ¸í´ë¦¬ì˜¤ ê°€ì¹˜ ë³€í™”")
print("=" * 70)
print(f"ì‹œì‘: {portfolio_value.index[0].strftime('%Y-%m')}")
print(f"ì¢…ë£Œ: {portfolio_value.index[-1].strftime('%Y-%m')}")
print(f"\nì´ˆê¸° íˆ¬ìê¸ˆ: 1.0")
print(f"ìµœì¢… ê°€ì¹˜: {portfolio_value.iloc[-1]:.4f}")
print(f"ì´ ìˆ˜ìµë¥ : {(portfolio_value.iloc[-1] - 1) * 100:.2f}%")

# ì—°í‰ê·  ìˆ˜ìµë¥  (CAGR)
years = len(portfolio_value) / 12
cagr = (portfolio_value.iloc[-1] ** (1/years) - 1) * 100
print(f"ì—°í‰ê·  ìˆ˜ìµë¥ (CAGR): {cagr:.2f}%")
````

## STEP 9: í¬íŠ¸í´ë¦¬ì˜¤ ì„±ê³¼ ì‹œê°í™”

````python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=portfolio_value.index,
    y=portfolio_value.values,
    mode='lines',
    name='í¬íŠ¸í´ë¦¬ì˜¤ ê°€ì¹˜',
    line=dict(color='#2E86AB', width=3),
    fill='tozeroy',
    fillcolor='rgba(46, 134, 171, 0.1)'
))

fig.add_hline(
    y=1.0,
    line_dash="dash",
    line_color="gray",
    annotation_text="ì´ˆê¸° íˆ¬ìê¸ˆ"
)

fig.update_layout(
    title='ì˜¬ì›¨ë” í¬íŠ¸í´ë¦¬ì˜¤ ëˆ„ì  ìˆ˜ìµë¥ ',
    xaxis_title='ë‚ ì§œ',
    yaxis_title='ê°€ì¹˜ (ì´ˆê¸°=1.0)',
    template='plotly_white',
    height=500
)

fig.show()
````

## STEP 10: ê°œë³„ ìì‚° ì„±ê³¼ ë¹„êµ

````python
# ê° ìì‚°ì˜ ëˆ„ì  ìˆ˜ìµ
cumulative_returns = (1 + monthly_return).cumprod()

fig = go.Figure()

# ê° ìì‚° ë¼ì¸
for col in cumulative_returns.columns:
    fig.add_trace(go.Scatter(
        x=cumulative_returns.index,
        y=cumulative_returns[col],
        mode='lines',
        name=col,
        line=dict(width=2)
    ))

# í¬íŠ¸í´ë¦¬ì˜¤ ì „ì²´ (êµµì€ ì„ )
fig.add_trace(go.Scatter(
    x=portfolio_value.index,
    y=portfolio_value.values,
    mode='lines',
    name='í¬íŠ¸í´ë¦¬ì˜¤ ì „ì²´',
    line=dict(color='black', width=4, dash='dash')
))

fig.update_layout(
    title='ìì‚°ë³„ ëˆ„ì  ìˆ˜ìµë¥  ë¹„êµ',
    xaxis_title='ë‚ ì§œ',
    yaxis_title='ëˆ„ì  ìˆ˜ìµ (ì´ˆê¸°=1.0)',
    template='plotly_white',
    height=600
)

fig.show()

# ìµœì¢… ìˆ˜ìµë¥  ë¹„êµ
final_returns = (cumulative_returns.iloc[-1] - 1) * 100
final_returns['í¬íŠ¸í´ë¦¬ì˜¤'] = (portfolio_value.iloc[-1] - 1) * 100

print("\nğŸ“Š ìì‚°ë³„ ìµœì¢… ìˆ˜ìµë¥ :")
print("=" * 70)
for asset, ret in final_returns.sort_values(ascending=False).items():
    print(f"{asset:20s}: {ret:+7.2f}%")
````

## STEP 11: ë¦¬ìŠ¤í¬ ì§€í‘œ ê³„ì‚°

````python
# ë³€ë™ì„± (ì—°ìœ¨í™”)
volatility = portfolio_return.std() * (12 ** 0.5) * 100

# ìƒ¤í”„ ë¹„ìœ¨ (ë¬´ìœ„í—˜ ìˆ˜ìµë¥  3% ê°€ì •)
risk_free_rate = 0.03
excess_return = cagr / 100 - risk_free_rate
sharpe_ratio = excess_return / (volatility / 100)

# ìµœëŒ€ ë‚™í­ (MDD)
cummax = portfolio_value.cummax()
drawdown = (portfolio_value - cummax) / cummax
max_drawdown = drawdown.min() * 100
mdd_date = drawdown.idxmin()

print("ğŸ“ˆ í¬íŠ¸í´ë¦¬ì˜¤ ì„±ê³¼ ì§€í‘œ")
print("=" * 70)
print(f"ì—°í‰ê·  ìˆ˜ìµë¥ (CAGR):     {cagr:>8.2f}%")
print(f"ì—°í‰ê·  ë³€ë™ì„±:           {volatility:>8.2f}%")
print(f"ìƒ¤í”„ ë¹„ìœ¨:              {sharpe_ratio:>8.2f}")
print(f"ìµœëŒ€ ë‚™í­(MDD):         {max_drawdown:>8.2f}%")
print(f"ìµœëŒ€ ë‚™í­ ë°œìƒ:          {mdd_date.strftime('%Yë…„ %mì›”')}")

print("\nğŸ’¡ í•´ì„:")
print(f"  â€¢ ì—°í‰ê·  {cagr:.1f}% ìˆ˜ìµ")
print(f"  â€¢ ë³€ë™ì„± {volatility:.1f}% (ìˆ˜ìµë¥ ì˜ ë“¤ì­‰ë‚ ì­‰í•¨)")

if sharpe_ratio > 1:
    evaluation = "í›Œë¥­í•¨ ğŸ‘"
elif sharpe_ratio > 0.5:
    evaluation = "ê´œì°®ìŒ ğŸ˜Š"
else:
    evaluation = "ë³´í†µ ğŸ˜"
    
print(f"  â€¢ ìƒ¤í”„ ë¹„ìœ¨ {sharpe_ratio:.2f} â†’ ìœ„í—˜ ëŒ€ë¹„ ìˆ˜ìµì´ {evaluation}")
print(f"  â€¢ ìµœì•…ì˜ ê²½ìš° {abs(max_drawdown):.1f}%ê¹Œì§€ í•˜ë½")
````

## STEP 12: ë‚™í­(Drawdown) ì‹œê°í™”

````python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=drawdown.index,
    y=drawdown.values * 100,
    mode='lines',
    name='Drawdown',
    fill='tozeroy',
    line=dict(color='red', width=2),
    fillcolor='rgba(255, 0, 0, 0.2)'
))

fig.update_layout(
    title='í¬íŠ¸í´ë¦¬ì˜¤ ë‚™í­(Drawdown) ì¶”ì´',
    xaxis_title='ë‚ ì§œ',
    yaxis_title='ë‚™í­ (%)',
    template='plotly_white',
    height=400
)

fig.show()
````

## STEP 13: ì›”ë³„ ìˆ˜ìµë¥  íˆíŠ¸ë§µ

````python
# ë°ì´í„° ì¤€ë¹„
monthly_ret_pct = portfolio_return * 100
monthly_ret_pct.index = pd.to_datetime(monthly_ret_pct.index)

heatmap_data = pd.DataFrame({
    'ì—°ë„': monthly_ret_pct.index.year,
    'ì›”': monthly_ret_pct.index.month,
    'ìˆ˜ìµë¥ ': monthly_ret_pct.values
})

heatmap_pivot = heatmap_data.pivot(index='ì—°ë„', columns='ì›”', values='ìˆ˜ìµë¥ ')

# íˆíŠ¸ë§µ ìƒì„±
fig = go.Figure(data=go.Heatmap(
    z=heatmap_pivot.values,
    x=[f"{m}ì›”" for m in heatmap_pivot.columns],
    y=heatmap_pivot.index,
    colorscale='RdYlGn',
    zmid=0,
    text=heatmap_pivot.values.round(2),
    texttemplate='%{text}%',
    textfont={"size": 10},
    colorbar=dict(title="ìˆ˜ìµë¥ (%)")
))

fig.update_layout(
    title='ì›”ë³„ ìˆ˜ìµë¥  íˆíŠ¸ë§µ',
    xaxis_title='ì›”',
    yaxis_title='ì—°ë„',
    height=400
)

fig.show()

print("\nğŸ’¡ ìƒ‰ê¹” ì˜ë¯¸:")
print("  ğŸŸ¢ ì´ˆë¡ìƒ‰: ìˆ˜ìµì´ ë‚œ ë‹¬")
print("  ğŸ”´ ë¹¨ê°„ìƒ‰: ì†ì‹¤ì´ ë‚œ ë‹¬")
print("  ğŸŸ¡ ë…¸ë€ìƒ‰: ê±°ì˜ ë³€ë™ ì—†ìŒ")
````
