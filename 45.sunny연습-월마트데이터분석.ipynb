```python
male_summary = (
    male_df_raw
    .groupby('Age')
    .agg(
        visit_count=('Purchase', 'count'),
        total_purchase=('Purchase', 'sum')
    )
    .reset_index()
)
```

```python
female_summary = (
    female_df_raw
    .groupby('Age')
    .agg(
        visit_count=('Purchase', 'count'),
        total_purchase=('Purchase', 'sum')
    )
    .reset_index()
)
```

```python
male_summary['avg_purchase'] = (
    male_summary['total_purchase'] / male_summary['visit_count']
)

female_summary['avg_purchase'] = (
    female_summary['total_purchase'] / female_summary['visit_count']
)
```

```python
top_age_df = (
    compare_df
    .sort_values('avg_purchase', ascending=False)
    .groupby('Gender')
    .head(1)
)

top_age_df
```

```python
compare_df['highlight'] = '기타'

for _, row in top_age_df.iterrows():
    mask = (
        (compare_df['Gender'] == row['Gender']) &
        (compare_df['Age'] == row['Age'])
    )
    compare_df.loc[mask, 'highlight'] = 'TOP'
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(
    data=compare_df,
    x='Age',
    y='avg_purchase',
    hue='highlight',
    palette={'TOP': '#D62728', '기타': '#D3D3D3'},
    errorbar=None
)

plt.title('Top Age Group Highlighted by Average Purchase')
plt.xlabel('Age Group')
plt.ylabel('Average Purchase per Visit')
plt.legend(title='구분')
plt.show()
```
