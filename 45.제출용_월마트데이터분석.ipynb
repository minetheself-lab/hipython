```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=compare_df,
    x='visit_count',
    y='avg_purchase',
    hue='Gender',
    s=120
)

plt.title('Visit Count vs Average Purchase by Age Group')
plt.xlabel('Visit Count')
plt.ylabel('Average Purchase per Visit')
plt.show()
```

```python
compare_df['purchase_efficiency'] = (
    compare_df['total_purchase'] / compare_df['visit_count']
)
```
