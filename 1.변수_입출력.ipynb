#변수의 영역 namespace
지역변수-특정영역, 전역변수-어디서나

```python
def local_exam(): #사용자정의 함수
    local_var=100
    print(f"함수내부: {local_var}")

local_exam()
local_var=200
local_exam()
print(f"함수외부:{local_var}")
```

```python
counter=0  #전역변수
def inc_counter():
    global counter
    #temp=a+1
    #a=temp
    #a=a+1
    #a += 1
    counter= counter+1
    print(counter)

inc_counter()
inc_counter()
inc_counter()
```
