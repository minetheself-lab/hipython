# 1변량 그래프 : 박스플롯

```python
tips_df
```

```python
# 여성의 청구서금액 vs. 팁
plt.boxplot(
    [tips_df[tips_df['sex']=='Female']['total_bill'], 
     tips_df[tips_df['sex']=='Female']['tip']],
    labels=['total_bill/Female', 'tip/Female']
)
plt.grid()
plt.show()
```

```python
# 남성의 청구서금액 vs. 팁
plt.boxplot(
    [tips_df[tips_df['sex']=='Male']['total_bill'], 
     tips_df[tips_df['sex']=='Male']['tip']],
    labels=['total_bill/Male', 'tip/Male']
)
plt.grid()
plt.show()
```

```python
tips_df.groupby(['day', 'time']).size()
```

```python
thur_lunch_tip = tips_df[
    (tips_df['day'] == 'Thur') & (tips_df['time'] == 'Lunch')
]['tip']

sat_dinner_tip = tips_df[
    (tips_df['day'] == 'Sat') & (tips_df['time'] == 'Dinner')
]['tip']

sun_dinner_tip = tips_df[
    (tips_df['day'] == 'Sun') & (tips_df['time'] == 'Dinner')
]['tip']
```

```python
plt.boxplot(
  [thur_lunch_tip, sat_dinner_tip, sun_dinner_tip],
  label=['Thur Lunch', 'Sat Dinner', 'Sun Dinner']
)
plt.show()
```

- 토요일 저녁에 팁 : 이상치 다수 > 전설적인 큰손
- 일요일 저녁의 변동성 : 박스가 크다 > 제각각, 
- 토,일 저녁의 팁의 금액이 대체로 크다 > 중앙값
