```python
# Regression Plot회귀차트의 약자, x,y값을 그래프에 먼저 찍어준다. -> 점들 사이를 관통하는 직선을 그려준다. ->신뢰구간: 색칠된 영역은 실제데이터가 이범위에 있을 가능성이 있다.
# seaborn은 부트스트랩 통계기법 사용한다. 95%신뢰구간, 가운데가 좁은 경우 추세선이 매우 정확하다는 뜻. 양끝이 넓으면 약간 불확실하다는 뜻이다.

sns.regplot(x=x, y=y)
```

```python
# regplot은 회귀차트이고, 기본적으로 산점도를 포함하고 있다. 그럼에도 불구하고 별도로 산점도 scatter를 지정하는 이유는 스타일을 변경하기 위해서다.
# scatter_kws는 Scatter Keyword Arguments의 약자. 점들만 따로 꾸미는 옵션,s는 사이즈, color는 색, line_kws는 Line Keyword Arguments의 약자, 가운데 지나는 회귀선 색깔
sns.regplot(
    x=df_tips['total_bill'],
    y=df_tips['tip'],
    scatter_kws={'s':30, 'color':'blue'},
    line_kws={'color':'red'}
)
```

```python
import folium
from math import radians, cos, sin, sqrt, atan2

# 출발지: 화랑로 51길 81
start_lat = 37.624938497387
start_lon = 127.08682064299

# 도착지: 노량진
end_lat = 37.514238
end_lon = 126.94204

# 지도 생성 (초기 중심은 두 지점의 평균)
m = folium.Map(
    location=[(start_lat + end_lat)/2, (start_lon + end_lon)/2],
    zoom_start=12
)

# 출발지 마커
folium.Marker(
    location=[start_lat, start_lon],
    popup="출발: 화랑로 51길 81",
    icon=folium.Icon(color='green')
).add_to(m)

# 도착지 마커
folium.Marker(
    location=[end_lat, end_lon],
    popup="도착: 노량진",
    icon=folium.Icon(color='red')
).add_to(m)

# 두 지점을 선으로 연결
folium.PolyLine(
    locations=[[start_lat, start_lon], [end_lat, end_lon]],
    color='blue'
).add_to(m)

# 거리 계산 (Haversine)
R = 6371  # 지구 반경(km)
dlat = radians(end_lat - start_lat)
dlon = radians(end_lon - start_lon)
a = sin(dlat/2)**2 + cos(radians(start_lat)) * cos(radians(end_lat)) * sin(dlon/2)**2
c = 2 * atan2(sqrt(a), sqrt(1-a))
distance_km = R * c

# 거리 표시 텍스트
folium.map.Marker(
    [(start_lat + end_lat)/2, (start_lon + end_lon)/2],
    icon=folium.DivIcon(html=f"<div style='font-size: 16pt; color: blue;'>거리: {distance_km:.2f} km</div>")
).add_to(m)

# 결과 저장
m.save("seoul_route.html")
```

```python
import folium
import requests

# 출발지 (화랑로 51길 81)
start_lat = 37.624938497387
start_lon = 127.08682064299

# 도착지 (노량진)
end_lat = 37.514238
end_lon = 126.94204

# OSRM API URL (자동차 기준)
url = (
    f"http://router.project-osrm.org/route/v1/driving/"
    f"{start_lon},{start_lat};{end_lon},{end_lat}"
    f"?overview=full&geometries=geojson"
)

# OSRM 요청
response = requests.get(url)
data = response.json()

# 경로 좌표 추출 (lon, lat)
route_coords = data["routes"][0]["geometry"]["coordinates"]

# (lon, lat) → (lat, lon) 변환 (Folium용)
route_latlon = [(lat, lon) for lon, lat in route_coords]

# 지도 생성 (중간 지점 기준)
m = folium.Map(
    location=[(start_lat + end_lat) / 2, (start_lon + end_lon) / 2],
    zoom_start=11
)

# 출발지 마커
folium.Marker(
    location=[start_lat, start_lon],
    popup="출발: 화랑로 51길 81",
    icon=folium.Icon(color="green")
).add_to(m)

# 도착지 마커
folium.Marker(
    location=[end_lat, end_lon],
    popup="도착: 노량진",
    icon=folium.Icon(color="red")
).add_to(m)

# 최적 경로 표시
folium.PolyLine(
    locations=route_latlon,
    color="blue",
    weight=5,
    opacity=0.8
).add_to(m)

# 저장
m.save("osrm_route.html")
```

```python
start = [37.624938, 127.086821]   # 화랑로51길 81
end   = [37.514575, 126.941778]   # 노량진

import requests

def get_route(start, end, profile):
    url = (
        f"http://router.project-osrm.org/route/v1/{profile}/"
        f"{start[1]},{start[0]};{end[1]},{end[0]}"
        f"?overview=full&geometries=geojson"
    )
    res = requests.get(url)
    data = res.json()
    
    coords = data["routes"][0]["geometry"]["coordinates"]
    return [[lat, lon] for lon, lat in coords]
```

```python
import folium

m = folium.Map(
    location=start,
    zoom_start=12
)
```

```python
car_layer = folium.FeatureGroup(name="🚗 자동차")

car_route = get_route(start, end, "driving")

folium.PolyLine(
    car_route,
    color="blue",
    weight=5
).add_to(car_layer)

car_layer.add_to(m)
```

```python
walk_layer = folium.FeatureGroup(name="🚶 도보")

walk_route = get_route(start, end, "walking")

folium.PolyLine(
    walk_route,
    color="green",
    weight=5,
    dash_array="5,5"
).add_to(walk_layer)

walk_layer.add_to(m)
```

```python
folium.Marker(start, popup="출발지").add_to(m)
folium.Marker(end, popup="도착지").add_to(m)
```
