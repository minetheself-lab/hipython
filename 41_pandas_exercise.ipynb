smoker 여부에 따라, group_val 컬럼에 '흡연자', '비흡연자' 표시하기
</div>

```python
df['group_val'] = '흡연자'
```

```python
df['group_val']=df['smoker'].map({'yes':'흡연자', 'no':'비흡연자'})
```

```python
df
```

merge()
df['smoker'] 기준
'yes' > '흡연자'
'no' > '비흡연자'

```python
#합치기 위한 테이블 정의
df_smoke = pd.DataFrame({'group_val':['흡연자', '비흡연자']}, index=['yes','no'])
df_smoke#.index
```

```python
df
```

```python
# 같은 이름의 컬럼이 있으면 _x, _y 로 자동으로 컬럼 이름을 바꾼다.
df_merge = pd.merge( df, df_smoke, left_on='smoker', right_index=True,  how='outer' )
df_merge
```
