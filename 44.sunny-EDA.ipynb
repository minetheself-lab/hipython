```python
# figure는 캔버스 사이즈 지정할 때 add_subplot은 fig의 메소드로 칸을생성해주는 메소드이다.괄호안에는 행개수, 열개수, 위치번호를 넣는다.
# 생성된 칸안에 히스토그램을 집어넣을 때 hist히스토그램 넣기, pct_carrier_delay 데이터 프레임 사용해서 20개 구간으로 나누고 테두리를 검정색으로 하고 plt.show()그래프 결과를 보여달라.
fig=plt.figure(figsize=(16,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

ax1.hist(df['pct_carrier_delay'], bins=20, edgecolor='black')
ax2.hist(df['pct_atc_delay'], bins=20, edgecolor='black')
ax3.hist(df['pct_weather_delay'], bins=20, edgecolor='black')
plt.show()
```

```python
#sns는 시본 라이브러리 별칭이다. pairplot은 데이터프레임안에 변수가 3개 있을 때 사용한다. 
# 상관분석에 해당, 왼쪽 상단에서 아래로 내려오는 대각선은 자기 자신과의 관계이고, 나머지는 다른 변수와의 관계이다.
# 산점도가 그려진다. 상관관계와 데이터 분포 상태를 한눈에 알 수 있다. 
sns.pairplot(df)
```

```python
# corr은 데이터프레임과 시리즈 객체에 특화된 메소드. 
# numeric_only=True: 데이터에 글자(문자열)가 섞여 있으면 계산이 안 되니까, "숫자로 된 것만 골라서 계산해!"
# 시본 라이브러리에 있는 히트맵 지도를 그리되, annot True는 칸 안에 실제 숫자(상관계수)를 적어달라는 뜻
corr_matrix=df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True)
```
